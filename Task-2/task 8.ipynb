```python
data1 = data[['date','total_cases','new_cases','total_deaths','new_tests','total_tests']]
data1.head()
```

```python
data1.date = pd.to_datetime(data.date)
data1.set_index('date', inplace = True)
```

```python
data1.head()
```

```python
data1.plot(figsize = (20 , 10), linewidth = 5, fontsize = 20,color='red')
plt.xlabel('Month', fontsize = 20)
```

```python
data1.new_cases.plot(figsize = (20 , 10), linewidth = 5, fontsize = 20,color='red')
plt.xlabel('Month', fontsize = 20)
```

```python
data1.total_deaths.plot(figsize = (20 , 10), linewidth = 5, fontsize = 20,color='red')
plt.xlabel('Month', fontsize = 20)
```

```python
data1.new_tests.plot(figsize = (20 , 10), linewidth = 5, fontsize = 20,color='red')
plt.xlabel('Month', fontsize = 20)
```

```python
data1.total_tests.plot(figsize = (20 , 10), linewidth = 5, fontsize = 20,color='red')
plt.xlabel('Month', fontsize = 20)
```

```python
total_cases = data1[['total_cases']]
total_cases.rolling(12).mean().plot(figsize = (20 , 10), linewidth = 5, fontsize = 20,color='red')
plt.xlabel('Month' , fontsize = 20);
```

```python
new_cases = data1[['new_cases']]
new_cases.rolling(12).mean().plot(figsize = (20 , 10), linewidth = 5, fontsize = 20,color='red')
plt.xlabel('Month' , fontsize = 20);
```

```python
total_deaths = data1[['total_deaths']]
total_deaths.rolling(12).mean().plot(figsize = (20 , 10), linewidth = 5, fontsize = 20,color='red')
plt.xlabel('Month' , fontsize = 20);
```

```python
new_tests = data1[['new_tests']]
new_tests.rolling(12).mean().plot(figsize = (20 , 10), linewidth = 5, fontsize = 20,color='red')
plt.xlabel('Month' , fontsize = 20);
```

```python
total_tests = data1[['total_tests']]
total_tests.rolling(12).mean().plot(figsize = (20 , 10), linewidth = 5, fontsize = 20,color='red')
plt.xlabel('Month' , fontsize = 20);
```

```python
total_cases = data1[['total_cases']]
total_cases.rolling(12).mean().plot(figsize = (20 , 10), linewidth = 5, fontsize = 20,color='red')
plt.xlabel('Month' , fontsize = 20);
```

```python
df_rm = pd.concat([total_cases.rolling(12).mean(), total_deaths.rolling(12).mean()])
df_rm.plot(figsize = (20 , 10), linewidth = 5, fontsize = 20,color='red')
plt.xlabel('Month' , fontsize = 20);
```

```python
total_cases.diff().plot(figsize = (20 , 10), linewidth = 5, fontsize = 20,color='red')
plt.xlabel('Month', fontsize = 20);
```
